```python
# 10. Make a scatterplot to represent  minmax_norm_ratings in function of the number of awards won by the book
#- Is there another representation that displays this in a more clear manner?
#- Optional: Can you plot a best fit linear regression line to represent the relationship?

def cor_award_minmax (df):
    sns.regplot(x = "award_count", y = "minmax_norm_ratings", line_kws={"color": "red"}, data = df)
    plt.xlim(0,20)
    plt.title('Correlation Between Awards Number and Min/Max Normalised Ratings', fontsize='12')
    plt.xlabel('Number of Awards')
    plt.ylabel('Min/Max Normalization')
    plt.show()
cor_award_minmax (df)


```

```python
# Correlation Between Page Numbers and Ranking ALL Books

def corr_page_rank(df):
    g=sns.regplot(x = "num_pages", y = "normalise_mean",line_kws={"color": "red"}, data = df, scatter_kws={'edgecolor':'white'})
    g.figure.set_size_inches(15,10)
    plt.xlim(-0,1000)
    plt.ylim(3.5,7)
    plt.title('Correlation Between Page Numbers and Ranking (All Books)', fontsize='20')
    plt.xlabel('Page numbers')
    plt.ylabel('Rank')
    plt.show()
    corr_pages_review = df.corr().loc["num_pages","normalise_mean"]
    print (f"Correlation between number of pages and rating is {round(corr_pages_review,4)}")
corr_page_rank(df)
```

```python
# Correlation Between Page Numbers and Ranking top 100 Books

top_100 = df.sort_values(by='normalise_mean', ascending=False)
top_100=  top_100[['title','normalise_mean', 'num_pages']] 
top_100 = top_100.head(100)

def corr_page_rank(df):
    g=sns.regplot(x = "num_pages", y = "normalise_mean",line_kws={"color": "red"}, data = top_100, scatter_kws={'edgecolor':'white'})
    g.figure.set_size_inches(15,10)
    #plt.xlim(-0,1000)
    #plt.ylim(3.5,7)
    plt.title('Correlation Between Page Numbers and Ranking (Top 100 ranked books)', fontsize='20')
    plt.xlabel('Page numbers')
    plt.ylabel('Rank')
    plt.show()
    corr_pages_review = top_100.corr().loc["num_pages","normalise_mean"]
    print (f"Correlation between number of pages and rating is {round(corr_pages_review,4)}")
corr_page_rank(df)
```
